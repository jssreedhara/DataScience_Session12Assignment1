```python
# Assignment 12 by Sreedhara Jagatagar  Sreenivasa
#2. Show me the average hours per week of all men who are working in private sector

#Develop the query
q_men_average=""" 
                SELECT SEX, WorkClass, avg(hours_per_week) 'Average Hours Per Week' 
                from adult_sqladb a 
                where ltrim(upper(a.sex)) ='MALE' 
                and ltrim(upper(a.WorkClass))='PRIVATE'
                group by Sex,WorkClass; 
            """
#
print(q_men_average)
pysqldf(q_men_average)
```

```python
# Assignment 12 by Sreedhara Jagatagar  Sreenivasa
#4. Are there any people who are married, working in private sector and having a masters degree

#Build the Query 
query_married_working_master=""" SELECT workclass, education,'Marrid',count(*)
                                 from adult_sqladb a 
                                 where ltrim(a.marital) in('Married-civ-spouse','Married-spouse-absent','Married-AF-spouse') 
                                 and ltrim(upper(a.WorkClass))='PRIVATE'
                                 and  ltrim(upper(a.education))='MASTERS'                                 
                                 group by a.workclass, a.education;
                            """

#print the query
print (query_married_working_master)

#Execute the SQL
pysqldf(query_married_working_master)
```

```python
# Assignment 12 by Sreedhara Jagatagar  Sreenivasa
#5. What is the average, minimum and maximum age group for people working in different sectors

#Build the Query
query_working =""" SELECT a.WorkClass,avg(a.Age),min(a.Age),max(a.Age)
                                 from adult_sqladb a
                                 group by a.WorkClass;
                            """
#print the query
print (query_working)

#execute the Query
pysqldf(query_working)
```

```python
# Assignment 12 by Sreedhara Jagatagar  Sreenivasa
#6. Calculate age distribution by country
#Build the query
query_age_company =""" SELECT a.native_country,a.Age, count(Age)
                                 from adult_sqladb a
                                 where ltrim(a.native_country)<>'?'
                                 group by a.native_country,a.Age;
                            """
#print the query
print (query_age_company)
#Execute the query
pysqldf(query_age_company)
```

```python
# Assignment 12 by Sreedhara Jagatagar  Sreenivasa
#7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

#Build the query
query_net_capital_loss =""" SELECT *,(a.capital_gain - a.capital_loss) as 'Net-Capital-Gain'
                                 from adult_sqladb a
                                 ;
                            """
#Print The Query
print (query_net_capital_loss)
#Execute the query
pysqldf(query_net_capital_loss)
```
